# CONDITIONAL STATEMENTS

Conditional statements, also known as conditional expressions or conditional structures, are fundamental constructs in programming that allow you to execute certain pieces of code based on whether a specified condition is true or false. These statements help control the flow of execution in a program.

### TYPES OF CONDITIONAL STATEMENTS

**1. 'if' statement**
- Executes a block of code only if a specified condition is true.

```python
a = 33
b = 50
if b < a:
    print("b is greater than a")
```

```python
x = 10
if x > 5:
    print("x is greater than 5")
```

```python
x = 10
y = (input("Enter a value for y: "))  
if y > 5:
    for i in range(y, x+1):
        print(i)  
```

**2.'if-else' Statement:**

-Executes one block of code if a condition is true, and another block of code if the condition is false.

```python
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

**3 .if-elif-else Statement:**

- Checks multiple conditions in sequence. If a condition is true, the corresponding block of code is executed. If none of the conditions are true, the else block is executed.

```python
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print('a is greater than b')
```

**4.Nested Conditional Statements:**

- A conditional statement inside another conditional statement.


```python
x =6
if x < 5:
    print("x is greater than 5")
    if x % 2 == 0:
        print("x is also even")
    else:
        print("x is odd")
else:
    print("x is 5 or less")
```

### LOGICAL OPERATORS

**5. And**

```python
a = 200
b = 33
c = 120
if a > b and c > a:
    print("Both conditions are True")
else:
    print("condition not satisfied")
```

**6.Or**

```python
a = 200
b = 33
c = 130
if a > b or a > c:
    print("At least one of the conditions is True")
else:
    print("hi")
```
