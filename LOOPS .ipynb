# Looping Statements
Execute a group of lines repeatedly.

Finite looping is the best practice in programming but infinite looping are done because they are easy at times.

1. For loop
2. While loop

Looping variables are generally named i, j, and k, while other names can be used.

#### 1. For loop:
General and most common finite looping.

For loop is used to iterate through a sequence or iterator.
Lists, tuple, dictionary and string.
Syntax:
```python
for value in iterable:  
    Statements
```

````python
# Example 1
l = [1, 2, 3, 4, 5]
for i in l:
    print(i)
````

````python
# Example 2
s = 'Hello'
for char in s:
    print(char)
````

````python
# Error for int
num = 10
for i in (num):
    print(num)
````

#### Range function:
Used to create an sequence/object from *start* to *stop* in *steps*

Syntax: 
```python
range(stop)  
range(start, stop, [step=1])
```

````python
#Example 2
for i in range(1, 7):
    print(i)
````

````python
print("1")
print("2")
print("3")
print("4")
print("5")
print("6")
````

````python
#Example 3
for i in range(0, 10, 2):
    print(i)
````

#### 2. While loops:
Most common for infinite looping.
Used to work on numbers or variable length iterables.
Preferred for variable conditions

Syntax: 
```python
while condition:  
    Statements
```

````python
# Example 2 using infinite looping
i = 1
while True:
    print(i)
    if i==1:
        break
    i+=1
````

````python
# Example 2 using infinite looping
i = 1
while True:
    print(i)
    if i==100:
        break
    i+=1
````

#### Loop Control statements:

Control the execution of the loop

1. Break
2. Continue
3. Pass

#### 1. break

This statement ends the loop when encountered.

#### 2. continue

Skips the loop execution when encountered.

#### 3. pass
No change to loop
Used when there no statements currently present

````python
i = 1
while i<5:
    j=0
    while True:
        print(i)
        j+=1 # j = j+1
        if j==4:
            break
    i+=1
````

````python
i = 1  # Initialize i to 1
while i < 6:  # Start a loop that continues as long as i is less than 6
    print(i)  # Print the current value of i
    if i == 3:  # Check if i is 3
        break  # Exit the loop if i is 3
    i += 1  # Increment i by 1
````

````python
#Example
i = 0
while i<11:
    i+=1
    if i==5: continue
    print(i)
````
